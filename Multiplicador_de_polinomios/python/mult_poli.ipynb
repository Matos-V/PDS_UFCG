```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi
from numpy.fft import fft, ifft
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['font.size'] = 30
plt.rcParams['figure.figsize'] = [30,14]
```

```python
bits = 8
N = 2**bits
x = np.arange(N)
time = np.arange(2*N-1)
y1 = np.zeros(2*N-1)
y2 = np.zeros(2*N-1)
y1[:N] = x
y2[:N] = x**2
result = y1*y2
```

```python
plt.plot(time,result)
plt.grid(True)
```

```python
Y1 = fft(y1)
Y2 = fft(y2)
Y3 = Y1*Y2
prod_poli = np.abs(ifft(Y3))
```

```python
normalizacao = (prod_poli[:N]/result[:N])[1:].max()
plt.plot(x,prod_poli[:N]/normalizacao,x,result[:N])
plt.grid(True)
plt.legend(['resultado - DFT','resultado - multiplicao'])
```

```python
cor = np.corrcoef(prod_poli[:N],result[:N])
```
